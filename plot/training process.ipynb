```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

peirl = [0.5313064435311811, 0.3069726319599036, 0.22615632398051026, 0.12040676251900106, 0.03623740962488502,
         0.03268597590075387, 0.029634376909485838, 0.02699257569929835, 0.024690114653618558, 0.02267107345981188,
         0.02089054730868073, 0.019312029891129862, 0.01790684003885212, 0.016652174673564793, 0.015527706133275308,
         0.014513707369430532, 0.009274004662601674, 0.0020811630795765372, 0.000684825211384468]

log_likelihood = [4.345     , 3.678     , 2.235     , 1.899     , 1.561     ,
       1.456     , 1.345     , 0.99389488, 0.59268354, 0.54683067,
       0.54682739, 0.52765286, 0.40065717, 0.39148799, 0.37046229,
       0.34651305, 0.19539403, 0.18415744]

mpl.rcParams['font.sans-serif'] = ['Times New Roman']
fig, ax1 = plt.subplots(figsize=(12, 5), dpi=1000)

color = 'tab:red'
ax1.set_xlabel('Epoch', fontsize=13, color='black', alpha=0.7)
ax1.set_ylabel('Reward difference between epochs', fontsize=13, color=color, alpha=0.7)
ax1.plot(peirl, 'r', marker='o', label='Reward Difference (PEMIRL)')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('Average Log Likelihood Ratio', fontsize=13, color=color, alpha=0.7)
ax2.plot(log_likelihood, 'b', marker='^', label='Average Log Likelihood')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper right')

x = np.arange(18)
y = np.linspace(0, 0.6, 10)
ax1.set_xticks(x)
ax1.set_yticks(y)
ax1.set_xlim(-0.3, 17.3)
ax1.grid(True, linestyle="--", alpha=0.5)

ax2.set_ylim(4.5, -0.5)

plt.tight_layout()
plt.savefig('reward_and_likelihood_plot.png', dpi=500, bbox_inches='tight')
plt.show()
```
