```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = './data/skeleton.csv'
data = pd.read_csv(file_path)
data=np.array(data)
```

```python
labels = [
    'Population density',
    'Land Use Mix',
    'Open Space Ratio',
    'Intersections',
    'Center',
    'Airport',
    'Railway',
    'Dock',
    'Coach',
    'Expressway',
    'Cycleway',
    'Suburban Road',
    'City Branch',
    'Inner Road',
    'Main Road',
    'Not Built Road',
    'SideWalk',
    'Urban Secondary',
    'Attractions',
    'Food&Beverages',
    'Tranportation',
    'Sport',
    'Public',
    'Enterprises',
    'Medical',
    'Government',
    'Life',
    'Education&Science',
    'Shopping',
    'Finance',
    'Male',
    'Female',
    'Young',
    'Middle',
    'Old',
    'Reward'
]
```

```python
correlation = pd.DataFrame(data = data,columns = labels,index=labels)
correlation=correlation.replace(0,np.nan)
```

```python
plt.style.use('ggplot')
rc = {'font.sans-serif': 'Times New Roman',
      'axes.unicode_minus': False}
sns.set_style(style='whitegrid',rc=rc)
plt.subplots(figsize = (30,20))

mask = np.zeros_like(data, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

cbar_kws = {
    "shrink": 0.7
}

sns.heatmap(correlation, 
            cmap=sns.diverging_palette(220,20,n=200),
            mask=mask,
            annot=True, 
            annot_kws={"fontsize":14},
            center = 0,
            cbar_kws=cbar_kws
           )

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig('./img/sns_causal.png',dpi=1000)
```
